```python
from efficientnet_pytorch import EfficientNet
import torch
import torch.nn as nn
from torchvision import transforms
from PIL import Image
import os

# Define the class labels
classes = ["Powdery mildew", "Healthy", "Cotton Boll Rot", "Bacterial blight", "Aphids"]

# Initialize the EfficientNet model with the correct number of output classes
model_test = EfficientNet.from_pretrained('efficientnet-b0', num_classes=8)

# Load the state dict from the checkpoint
checkpoint_path = "first_tuned_model_efficientnet_cotton_disease.pth"
checkpoint = torch.load(checkpoint_path)
model_dict = model_test.state_dict()

# Filter out unnecessary keys
checkpoint = {k: v for k, v in checkpoint.items() if k in model_dict}

# Update model's state_dict
model_dict.update(checkpoint)

# Load the modified state_dict
model_test.load_state_dict(model_dict)

# Modify the architecture to match the desired number of output classes (5)
model_test._fc = nn.Linear(model_test._fc.in_features, len(classes))

# Optional: Transfer the model to the GPU
# device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# model_test = model_test.to(device)

# Set the model to evaluation mode
model_test.eval()

# Define transformations for input images
transform = transforms.Compose([
#     transforms.Resize((64, 64)),
    transforms.ToTensor(),
])

# Loop over the folder of the testing dataset
testing_folder = "G:\\NAVANEETH\\#Cotton_Leaf_Disease_Detection\\Dataset2\\Cotton leaf_Validation\\validation\\Cotton plant disease-Validation\\Cotton plant disease-Validation\\Bacterial Blight edited"
for image_name in os.listdir(testing_folder):
    if image_name.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp')):
        # Load the image
        image_path = os.path.join(testing_folder, image_name)
        image = Image.open(image_path).convert("RGB")

        # Apply transformations
        image = transform(image)

        # Add batch dimension
        image = image.unsqueeze(0)

        # Make prediction
        with torch.no_grad():
            output = model_test(image)

        # Get predicted class
        _, predicted_idx = torch.max(output, 1)
        predicted_class = classes[predicted_idx.item()]

        print(f"Image: {image_name}, Predicted Class: {predicted_class}")
```
